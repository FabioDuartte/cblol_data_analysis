# 1 - O que é League Of Legends?

# League of Legends

League of Legends é um jogo online multiplayer de batalha de arena (MOBA) desenvolvido e publicado pela Riot Games. Lançado em 2009, o jogo se tornou um dos mais populares no gênero, com uma base de jogadores enorme e uma cena de esportes eletrônicos muito ativa. Em League of Legends, os jogadores assumem o papel de "campeões", personagens com habilidades únicas, e competem em equipes para destruir a base inimiga enquanto defendem a sua própria. O jogo é conhecido por sua jogabilidade estratégica, a necessidade de trabalho em equipe e a constante evolução do meta jogo com atualizações frequentes.

## Mapa e Objetivo Principal

O jogo é jogado em um mapa chamado Summoner's Rift, que é dividido em três rotas principais, top, mid e bot, conectadas por uma selva. O objetivo principal é destruir a base inimiga, chamada de "Nexus", localizada na extremidade oposta do mapa. Para chegar ao Nexus inimigo, os jogadores precisam derrubar torres, inibidores e enfrentar as defesas da equipe adversária.

## Escolha de Campeões

Antes de cada partida, os jogadores selecionam seus campeões de um grande pool disponível, cada um com habilidades únicas e papéis específicos. Existem campeões de dano, tanques, suportes, assassinos, entre outros. A composição da equipe e a escolha dos campeões são fundamentais para o sucesso em uma partida.

## Fases do Jogo

- **Fase de Rotação:** Os jogadores começam nas rotas designadas, onde lutam contra os minions inimigos e tentam derrubar as torres adversárias enquanto defendem as suas.
- **Fase de Selva:** Jogadores responsáveis pela selva, chamados de junglers, percorrem a área entre as rotas, derrotam monstros neutros e ajudam as rotas com ganks (ataques surpresa) para obter vantagem.
- **Fase de Equipe:** À medida que o jogo avança, as equipes se agrupam e lutam por objetivos importantes, como dragões, barões e torres inibidoras. O trabalho em equipe e a coordenação são essenciais durante esta fase.

## Evolução dos Campeões

Os campeões ganham experiência e ouro ao derrotar minions, monstros neutros e inimigos. Com o ouro, eles podem comprar itens para melhorar suas estatísticas e habilidades. A evolução dos campeões ao longo da partida é crucial para se tornarem mais poderosos e eficazes em combate.

## Estratégia e Coordenação

League of Legends é profundamente estratégico. A comunicação e a coordenação entre os membros da equipe são vitais para garantir o sucesso. Decisões táticas, como quando atacar, quando recuar, quais objetivos priorizar e como posicionar-se no mapa, podem determinar o resultado de uma partida.

## Esports e Comunidade

League of Legends tem uma comunidade enorme e ativa em todo o mundo. Além do aspecto competitivo nas filas ranqueadas do jogo, há também uma cena profissional de esportes eletrônicos extremamente desenvolvida, com torneios de prestígio como o Campeonato Mundial de League of Legends (Worlds).

## Wards (Sentinelas)

No League of Legends, as "wards" são dispositivos colocados no mapa para fornecer visão a uma equipe. Elas desempenham um papel fundamental no controle do mapa e na prevenção de emboscadas, permitindo que os jogadores vejam áreas do mapa que normalmente estariam fora de sua visão direta. Existem dois tipos principais de wards:

**Wards Normais**: Estas são as wards básicas que os jogadores podem comprar no jogo. Elas têm um tempo de vida limitado e podem ser colocadas em locais estratégicos para fornecer visão temporária em uma área específica. As wards normais são frequentemente usadas para monitorar pontos de acesso importantes, como entradas da selva ou objetivos neutros, como o Dragão ou o Barão.

**Wards de Controle (Control Wards)**: As wards de controle, às vezes chamadas de "pink wards", são wards especiais que duram indefinidamente, a menos que sejam destruídas pelos jogadores inimigos. Além de fornecer visão em uma área, as wards de controle também revelam e desativam as wards inimigas próximas. Isso as torna extremamente valiosas para negar a visão inimiga e manter o controle do mapa. As wards de controle são frequentemente usadas para estabelecer domínio sobre objetivos importantes e áreas-chave do mapa.

# 5 - Análise dos Jogadores 

### 5.1 - Jogador que com o melhor kda

```python
# Criando a coluna kda e calculando para cada jogador
cblol['kda'] = np.where(cblol['deaths'] == 0, np.nan, (cblol['kills'] + cblol['assists']) / cblol['deaths'])

# Agrupando por equipe e encontrar o jogador com o melhor KDA em cada equipe
best_kda_players = cblol.groupby('teamname').apply(lambda x: x.loc[x['kda'].idxmax()])

# Ordenando os resultados em ordem decrescente pelo KDA
best_kda_players = best_kda_players.sort_values(by='kda', ascending=False)

# jogador com o melhor KDA em cada equipe
print(best_kda_players[['playername', 'position', 'kda']])
```

```python
plt.figure(figsize=(10, 8))
plt.pie(best_kda_players['kda'], labels=best_kda_players['playername'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição do KDA dos Jogadores com Melhor KDA por Equipe')
plt.axis('equal')  # Garante que o gráfico de pizza seja desenhado como um círculo
plt.show()
```

### 6.3 - Presença dos campeões entre picks e bans

```python
# Combinando as colunas de escolhas e bans em uma única série
escolhas = cblol['champion']
bans = pd.concat([cblol['ban1'], cblol['ban2'], cblol['ban3'], cblol['ban4'], cblol['ban5']])
presenca_total = pd.concat([escolhas, bans])

# Usando o value_counts() para contar a frequência de cada campeão
contagem_presenca_total = presenca_total.value_counts()

# Calculando a porcentagem de presença de cada campeão
porcentagem_presenca = (contagem_presenca_total / len(presenca_total)) * 100

# Calculando a porcentagem total de presença de todos os campeões para normalização
porcentagem_total = porcentagem_presenca.sum()

# Exibindo a porcentagem de presença dos campeões
print(porcentagem_presenca.round(2).astype(str) + '%')
print("\nPorcentagem total de presença de todos os campeões:", round(porcentagem_total, 2), '%')
```

```python
# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
porcentagem_presenca.plot(kind='bar')
plt.xlabel('Campeões')
plt.ylabel('Porcentagem de Presença')
plt.title('Porcentagem de Presença dos Campeões entre Picks e Bans')
plt.xticks(rotation=90)  # Rotacionar os rótulos do eixo x para melhorar a legibilidade
plt.grid(axis='y')  # Adicionar grade apenas no eixo y para melhorar a legibilidade
plt.tight_layout()  # Ajustar layout para evitar sobreposição de texto
plt.show()
```
